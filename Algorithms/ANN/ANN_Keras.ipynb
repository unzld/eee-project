```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle

df = pd.read_csv('d3_avg_error_adjustedloads.csv')
df = df.astype({col: 'float32' for col in df.select_dtypes('int64').columns})
df = df.drop(df.filter(regex='power').columns, axis=1)
df['id']=np.divmod(np.arange(len(df)),2)[0]+1
df = shuffle(df)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GroupShuffleSplit
import numpy as np

#train, test, y_train, y_test = train_test_split(df, df['label'], test_size=0.20, stratify=df['label'], random_state=100)
#x_train = train.drop(columns=['label', 'category', 'id'])
#x_test = test.drop(columns=['label', 'category', 'id'])

train_inds, test_inds = next(GroupShuffleSplit(test_size=0.20, n_splits=2, random_state=100).split(df, groups=df['id']))
x_train = df.iloc[train_inds].drop(columns=['label', 'category', 'id'])
x_test = df.iloc[test_inds].drop(columns=['label', 'category', 'id'])
y_train = df.iloc[train_inds]['label']
y_test = df.iloc[test_inds]['label']

#x_train = shuffle(x_train, random_state=100)
#y_train = shuffle(y_train, random_state=100)
#x_test = shuffle(x_test, random_state=200)
#y_test = shuffle(y_test, random_state=200)
```

```python
# Scale the train set
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#Standardization
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)

# Normalization
norm = MinMaxScaler()
x_train_norm = norm.fit_transform(x_train)
x_test_norm = norm.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
from keras.optimizers import SGD

def create_model(lr=0.01, nodes=1, activation='relu'):
	# create model
  # define the keras model
  model = Sequential()
  model.add(Dense(nodes, input_dim=63, activation=activation)) #input
  model.add(Dense(nodes, activation=activation))
  model.add(Dense(nodes, activation=activation))
  model.add(Dense(1, activation='softmax')) #output
  optimizer = Adam(learning_rate=lr)
  model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
  return model
```

```python
from sklearn.model_selection import GridSearchCV

#model = KerasClassifier(build_fn=create_model, batch_size=256, epochs=100, verbose=1)
#activation = ['relu']
#learn_rate = 10.0 ** -np.arange(1, 7)
#nodes = [12]
#param_grid = dict(lr=learn_rate, nodes=nodes, activation=activation)
#grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1, cv=10, verbose=1)
#grid_result = grid.fit(x_train_norm, y_train)

lr = 0.1
nodes = 60
epoch = 100
batch_size = 32

model = Sequential()
model.add(Dense(nodes, input_dim=8, activation='relu')) #input
model.add(Dense(nodes, activation='relu'))
model.add(Dense(nodes, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #output

optimizer = SGD(learning_rate=lr)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
model.fit(x_train_scaled, y_train, verbose=1, epochs=epoch, batch_size=batch_size)
```

```python
#print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
_, accuracy_train = model.evaluate(x_train_scaled, y_train)
print('Train Accuracy: %.2f' % (accuracy_train*100))

_, accuracy_test = model.evaluate(x_test_scaled, y_test)
print('Test Accuracy: %.2f' % (accuracy_test*100))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

y_pred = model.predict_classes(x_test_scaled)
y_pred = y_pred[:, 0]
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
acc = (tp + tn) / (tn + fp + fn + tp)
pr = tp / (tp + fp)
dr = tp / (tp + fn)
f1 = (2*dr*pr) / (dr + pr)

print('True positives: {}'.format(tp))
print('True negatives: {}'.format(tn))
print('False positives: {}'.format(fp))
print('False negatives: {}'.format(fn))
```
